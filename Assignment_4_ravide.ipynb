```
# File management
import os # Load library os dari Python

print(os.listdir('.')) # List semua file di directory sekarang (.)

print(os.listdir('./sample_data')) # List file di directory sample_data
```

```
# Membaca file

# Cek apakah file ada atau tidak?

directory = os.path.join('sample_data', 'california_housing_train.csv') # Output: sample_data/california_housing_train.csv

print(directory)

if os.path.isfile(directory):
  print("File ditemukan")

text_file = open(directory, "r")

# print(text_file.read())

lines = text_file.readlines()

# print(lines)

print("Jumlah data (original) ", len(lines))

for line in lines[:3]: # Print 3 baris pertama
  print(line)

  # line = line.strip()

  split_columns = line.split(",")

  print(split_columns)

parsed_values = []

for index, line in enumerate(lines):
  split_columns = line.split(",")

  if index == 0:
    continue
  
  parsed_values.append([float(split_columns[0]), float(split_columns[1]), float(split_columns[2]), float(split_columns[3]), float(split_columns[4]), float(split_columns[5]), float(split_columns[6]), float(split_columns[7]), float(split_columns[8])])

print("Jumlah data (parsed) ", len(parsed_values))

print(parsed_values)
```

```
# Jawaban:

# menyimpan data masing-masing feature ke dalam sebuah list
list_housing_median_age = []
list_total_rooms = []
list_total_bedrooms = []
list_population = []
list_households = []
list_median_income = []
list_median_house_value = []

for list_col in parsed_values:
  for indek, value in enumerate(list_col):
    if indek == 0:
      continue
    if indek == 1:
      continue
    if indek == 2:
      list_housing_median_age.append(value)
    if indek == 3:
      list_total_rooms.append(value)
    if indek == 4:
      list_total_bedrooms.append(value)
    if indek == 5:
      list_population.append(value)
    if indek == 6:
      list_households.append(value)
    if indek == 7:
      list_median_income.append(value)
    if indek == 8:
      list_median_house_value.append(value)

# melihat data masing-masing feature
print(list_housing_median_age)
print(list_total_rooms)
print(list_total_bedrooms)
print(list_population)
print(list_households)
print(list_median_income)
print(list_median_house_value)
```

```
# melihat informasi jumlah data tiap feature
print(len(list_housing_median_age))
print(len(list_total_rooms))
print(len(list_total_bedrooms))
print(len(list_population))
print(len(list_households))
print(len(list_median_income))
print(len(list_median_house_value))
```

```
# membuat fungsi informasi statistik deskriptif pada data 
def func_all(list_colmns):
  n = len(list_colmns)
  mean_list_colmns = sum(list_colmns) / len(list_colmns)
  std_list_colmns = (sum([(item - mean_list_colmns)**2 for item in list_colmns]) / (n - 1)) ** 0.5
  min_list_colmns = min(list_colmns)
  max_list_colmns = max(list_colmns)
  q1_list_colmns = sorted(list_colmns)[round(0.25*(n-1))]
  q2_list_colmns = sorted(list_colmns)[round(0.5*(n-1))]
  q3_list_colmns = sorted(list_colmns)[round(0.75*(n-1))]

  # mencetak nilai informasi statistik deskriptif data
  print('Nilai Mean \t:', mean_list_colmns)
  print('Nilai STD \t:', std_list_colmns)
  print('Nilai Min \t:', min_list_colmns)
  print('Nilai Max \t:', max_list_colmns)
  print('Nilai Q1 \t:', q1_list_colmns)
  print('Nilai Q2 \t:', q2_list_colmns)
  print('Nilai Q3 \t:', q3_list_colmns)

  # menentukan data ada yang kosong/tidak
  if n < 17000:
    print('Data ada yang kosong')
  else:
    print('Data tidak ada yang kosong')

# Memanggil informasi Data Statistik Deskriptif Tiap Kolom
nama_colmns = ["housing_median_age","total_rooms","total_bedrooms","population","households","median_income","median_house_value"]
list_colmns = [list_housing_median_age, list_total_rooms, list_total_bedrooms, list_population, list_households, list_median_income, list_median_house_value]

for i in range(len(nama_colmns)):
  print(f'Informasi pada kolom {nama_colmns[i]}: ')
  func_all(list_colmns[i])
  print('\n')
```
